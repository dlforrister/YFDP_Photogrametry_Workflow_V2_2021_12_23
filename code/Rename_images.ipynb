```python
import os
import pandas as pd
import PIL
from PIL import Image
from PIL import ImageStat
import glob
from datetime import datetime
import shutil
```

```python
def rename(baseDir):
    failed_photos = pd.DataFrame(columns={'filename': [], 'flight': [], 'n': [],'message': []})
    if not os.path.exists(baseDir + '/Proccessed_Images_V2/'):
        os.makedirs(baseDir + '/Proccessed_Images_V2/')
    if not os.path.exists(baseDir + "/Proccessed_Images_V2/Phantom/"):
        os.makedirs(baseDir + "/Proccessed_Images_V2/Phantom/")
    i = 1
    for flight in os.listdir(baseDir + "/Phantom/"):
      if flight == os.listdir(baseDir + "/Phantom/")[0]:
          x=1
          print(baseDir + "/Phantom" + "/" + flight+ "/*.JPG")
          file_list = glob.glob(baseDir + "/Phantom" + "/" + flight + "/*.JPG")
          file_list.sort()
          #file_list.sort(key=os.path.getmtime)
          for filename in file_list:
              try:
                  pic = PIL.Image.open(filename)
                  pic_date = datetime.strptime(str(pic._getexif()[36867]), '%Y:%m:%d %H:%M:%S')
                  shutil.copy(filename, baseDir + "/Proccessed_Images_V2/Phantom/" + pic_date.strftime("%Y") + "_" + pic_date.strftime('%b') + "_" + pic_date.strftime('%d') + "_Phantom_Flight_" + str(i) + "_" + str(x) + ".JPG")
                  if x % 10 == 0:
                    print(x)
                  pic.close()
                  x+=1
              except IOError as error:
                  failed_photos = failed_photos.append({'file_name': filename,'flight': flight, 'n' : x,'message': error},ignore_index=True)
                  print('skipping {}'.format(x))
                  x+=1
                  continue
          i+=1
```

```python
rename(baseDir = "/Volumes/GoogleDrive/My Drive/2018_Yasuni_Drone_Mapping/7_2019_Mar/Feb_28_2019/") 
```
